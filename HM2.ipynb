```python
def rgbtohsv(image):
    new_img = np.zeros(image.shape, image.dtype)
    for x in range(image.shape[0]):
        for y in range(image.shape[1]):
                r = img[x,y,0]/255
                g = img[x,y,1]/255
                b = img[x,y,2]/255
                maxx = max(r,g,b)
                minn = min(r,g,b)
                delta = maxx - minn

                if delta == 0:
                    h = 0
                elif maxx == r:
                    h = (60 * ((g-b)/delta) + 360) % 360
                elif maxx == g:
                    h = (60 * ((b-r)/delta) + 120) % 360
                elif maxx == b:
                    h = (60 * ((r-g)/delta) + 240) % 360
                v = maxx
                if maxx == 0:
                    s = 0
                else:
                    s = delta/maxx

                new_img[x,y] = [(h/2), (s*255), (v*255)]

    return new_img
```

```python
#prima functie facuta
def rgbtohsv2(r, g, b):
    r /=255
    g /=255
    b /=255
    maxx = max(r,g,b)
    minn = min(r,g,b)
    delta = maxx - minn
    
    if delta == 0:
        h = 0
    elif maxx == r:
        h = (60 * ((g-b)/delta) + 360) % 360
    elif maxx == g:
        h = (60 * ((b-r)/delta) + 120) % 360
    elif maxx == b:
        h = (60 * ((r-g)/delta) + 240) % 360
    v = maxx
    if maxx == 0:
        s = 0
    else:
        s = delta/maxx
    return h, s, v
```

```python
gaussian = cv2.GaussianBlur(img, (3,3), 0)

io.imshow(gaussian)
```

```python
median = cv2.medianBlur(img,ksize = 3)
io.imshow(median)
```

```python
bilateral = cv2.bilateralFilter(img, 3, 75, 75)
io.imshow(bilateral)
```

```python
grayimg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
graymedian = cv2.cvtColor(median, cv2.COLOR_BGR2GRAY)
graygaussian = cv2.cvtColor(gaussian, cv2.COLOR_BGR2GRAY)
graybilateral = cv2.cvtColor(bilateral, cv2.COLOR_BGR2GRAY)
```

```python
io.imshow(graymedian)
```

```python
from skimage.metrics import structural_similarity as ssim
```

```python
ssim1 = ssim(grayimg, graymedian,
                    data_range = grayimg.max() - grayimg.min(), multichannel=True)
ssim2 = ssim(grayimg, graygaussian,
                    data_range = grayimg.max() - grayimg.min(), multichannel=True)
ssim3 = ssim(grayimg, graybilateral,
                    data_range = grayimg.max() - grayimg.min(), multichannel=True)
```

```python
plt.figure(1)
    
fig = plt.figure(figsize=(18, 9))

fig.add_subplot(1, 4, 1)
plt.title('Original')
plt.xlabel('1.0')
plt.imshow(img)

fig.add_subplot(1, 4, 2)
plt.title('Median Filtered')
plt.xlabel(ssim1)
plt.imshow(median)

fig.add_subplot(1, 4, 3)
plt.title('Gaussian Filtered')
plt.xlabel(ssim2)
plt.imshow(gaussian)

fig.add_subplot(1, 4, 4)
plt.title('Bilateral Filtered')
plt.xlabel(ssim3)
plt.imshow(bilateral)

#plt.subplots_adjust(right=1, wspace=0, hspace=0.3)
plt.show()
```
